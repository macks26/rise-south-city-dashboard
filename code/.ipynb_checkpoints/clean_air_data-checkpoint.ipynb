```python
# Script to clean Clarity hourly data
import pandas as pd

def clean_clarity_data(hourly_df, location_file_path, output_file_path):
    # Load the datasets
    location_df = pd.read_csv(location_file_path)

    # Join the two dataframes on the ID column
    joined_df = pd.merge(hourly_df, location_df, left_on='datasourceId', right_on='Datasource ID')
    joined_df.drop(columns=['datasourceId'], inplace=True)

    # Identify NO2 and PM2.5 columns
    no2_columns = [col for col in joined_df.columns if 'no2' in col.lower()]
    pm25_columns = [col for col in joined_df.columns if 'pm2_5' in col.lower()]

    # Filter the dataframe to include only the columns of interest
    joined_df = joined_df[['startOfPeriod', 'Name', 'Datasource ID', 'locationLatitude', 'locationLongitude']
                           + no2_columns + pm25_columns]

    # Convert the 'startOfPeriod' column to datetime
    joined_df['startOfPeriod'] = pd.to_datetime(joined_df['startOfPeriod'])

    # Update the NO2 and PM2.5 columns to include only relevant ones
    no2_columns = [col for col in no2_columns if 'value' in col and 'WaDwer' not in col]
    pm25_columns = [col for col in pm25_columns if 'value' in col and 'WaDwer' not in col and 'Mass' in col]

    # Filter dataframe with updated columns and location/time columns
    joined_df = joined_df[['startOfPeriod', 'Name', 'Datasource ID', 'locationLatitude', 'locationLongitude']
                           + no2_columns + pm25_columns]

    # Rename columns to more readable format
    joined_df.rename(columns={
        'startOfPeriod': 'time',
        'Name': 'location_name',
        'Datasource ID': 'location_id',
        'locationLatitude': 'latitude',
        'locationLongitude': 'longitude',
        'no2Conc1HourMean.value': 'no2_1h_mean',
        'no2Conc1HourMeanUsEpaAqi.value': 'no2_1h_mean_aqi',
        'pm2_5ConcMass1HourMean.value': 'pm2_5_1h_mean',
        'pm2_5ConcMassNowcastUsEpaAqi.value': 'pm2_5_1h_mean_aqi',
        'pm2_5ConcMass24HourRollingMean.value': 'pm2_5_24h_mean',
    }, inplace=True)

    # Drop unnecessary columns
    if 'pm2_5_nowcast' in joined_df.columns:
        joined_df.drop(columns=['pm2_5_nowcast'], inplace=True)

    # Impute missing values using forward fill followed by backward fill
    joined_df.ffill(inplace=True)
    joined_df.bfill(inplace=True)

    # Ensure latitude and longitude are in the correct format
    joined_df['longitude'] = joined_df['longitude'].apply(lambda x: -abs(x))
    joined_df['latitude'] = joined_df['latitude'].apply(lambda x: abs(x))

    # Save the cleaned dataframe to a new CSV file
    joined_df.to_csv(output_file_path, index=False)
```

```python
# Example usage
april_df = pd.read_csv('../data/risesouthcity_april_hourly.csv')

# Drop unnecessary columns
april_df.drop(columns=['Name'], inplace=True)
april_df.head()
```

```python
concat_df = pd.concat([hourly_df[april_df.columns], april_df], ignore_index=True)
concat_df.head()
```

```python
location_file_path = '../data/risesouthcity_locations.csv'
output_file_path = '../data/clarity_cleaned_hourly_data.csv'

# Clean the data
clean_clarity_data(concat_df, location_file_path, output_file_path)

# Check the cleaned data
cleaned_df = pd.read_csv(output_file_path)
cleaned_df.head()
```

## **PurpleAir Monitors**

```python
# AQI calculation
def calculate_pm2_5_aqi(C_p):
    if pd.isna(C_p):
        return None

    C_p = float(str(C_p)[:str(C_p).find('.')+2]) if '.' in str(C_p) else float(C_p)

    breakpoints = [
        (0.0,   9.0,   0,   50),
        (9.1,   35.4,  51,  100),
        (35.5,  55.4,  101, 150),
        (55.5,  125.4, 151, 200),
        (125.5, 225.4, 201, 300),
        (225.5, 500.4, 301, 500)
    ]

    for BP_Lo, BP_Hi, I_Lo, I_Hi in breakpoints:
        if BP_Lo <= C_p <= BP_Hi:
            I_p = ((I_Hi - I_Lo) / (BP_Hi - BP_Lo)) * (C_p - BP_Lo) + I_Lo
            return round(I_p)

    return None

def clean_purpleair_data(df):
    # Process PurpleAir API data
    df = df.rename(columns={
        'time_stamp': 'time',
        'sensor_name': 'location_name',
        'sensor_index': 'location_id',
        'pm2.5_atm': 'pm2_5_1h_mean',
        'temperature': 'temp',
        'humidity': 'rh'
    })

    df['time'] = pd.to_datetime(df['time'], utc=True)
    df['time'] = df['time'].dt.tz_convert('US/Pacific').dt.tz_localize(None)

    # Add date column for grouping
    df['date'] = df['time'].dt.date

    # Calculate rolling 24-hour mean per sensor
    df['pm2_5_24h_mean'] = (
        df.groupby('location_id')['pm2_5_1h_mean']
        .transform(lambda x: x.rolling(window=24, min_periods=1).mean())
    )

    # AQI Calculations for rolling 24-hour mean
    df['pm2_5_1h_mean_aqi'] = df['pm2_5_24h_mean'].apply(calculate_pm2_5_aqi)

    # Impute missing values for the entire dataframe
    df.ffill(inplace=True)
    df.bfill(inplace=True)

    # Column arrangement
    df = df[['time', 'location_name', 'location_id', 'latitude', 'longitude',
                            'pm2_5_1h_mean', 'pm2_5_24h_mean', 'pm2_5_1h_mean_aqi']]

    # Sort and save
    df = df.sort_values('time')
    df.to_csv("purpleair_api_cleaned_data.csv", index=False)
```

```python
purple_df = pd.read_csv('../data/uncleaned_purple_air_sensor_data.csv')

# Clean the PurpleAir data
clean_purpleair_data(purple_df)

# Check the cleaned PurpleAir data
purple_df_cleaned = pd.read_csv("purpleair_api_cleaned_data.csv")
purple_df_cleaned.head()
```
