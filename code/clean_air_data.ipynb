```python
# Script to clean Clarity hourly data
import pandas as pd

def clean_clarity_data(hourly_file_path, location_file_path, output_file_path):
    # Load the datasets
    hourly_df = pd.read_csv(hourly_file_path)
    location_df = pd.read_csv(location_file_path)

    # Join the two dataframes on the ID column
    joined_df = pd.merge(hourly_df, location_df, left_on='datasourceId', right_on='Datasource ID')
    joined_df.drop(columns=['datasourceId'], inplace=True)

    # Identify NO2 and PM2.5 columns
    no2_columns = [col for col in joined_df.columns if 'no2' in col.lower()]
    pm25_columns = [col for col in joined_df.columns if 'pm2_5' in col.lower()]

    # Filter the dataframe to include only the columns of interest
    joined_df = joined_df[['startOfPeriod', 'Name', 'Datasource ID', 'locationLatitude', 'locationLongitude']
                           + no2_columns + pm25_columns]

    # Convert the 'startOfPeriod' column to datetime
    joined_df['startOfPeriod'] = pd.to_datetime(joined_df['startOfPeriod'])

    # Update the NO2 and PM2.5 columns to include only relevant ones
    no2_columns = [col for col in no2_columns if 'value' in col and 'WaDwer' not in col]
    pm25_columns = [col for col in pm25_columns if 'value' in col and 'WaDwer' not in col and 'Mass' in col]

    # Filter dataframe with updated columns and location/time columns
    joined_df = joined_df[['startOfPeriod', 'Name', 'Datasource ID', 'locationLatitude', 'locationLongitude']
                           + no2_columns + pm25_columns]

    # Rename columns to more readable format
    joined_df.rename(columns={
        'startOfPeriod': 'time',
        'Name': 'location_name',
        'Datasource ID': 'location_id',
        'locationLatitude': 'latitude',
        'locationLongitude': 'longitude',
        'no2Conc1HourMean.value': 'no2_1h_mean',
        'no2Conc1HourMeanUsEpaAqi.value': 'no2_1h_mean_aqi',
        'pm2_5ConcMass1HourMean.value': 'pm2_5_1h_mean',
        'pm2_5ConcMassNowcastUsEpaAqi.value': 'pm2_5_1h_mean_aqi',
        'pm2_5ConcMass24HourRollingMean.value': 'pm2_5_24h_mean',
    }, inplace=True)

    # Drop unnecessary columns
    if 'pm2_5_nowcast' in joined_df.columns:
        joined_df.drop(columns=['pm2_5_nowcast'], inplace=True)

    # Impute missing values using forward fill followed by backward fill
    joined_df.ffill(inplace=True)
    joined_df.bfill(inplace=True)

    # Save the cleaned dataframe to a new CSV file
    joined_df.to_csv(output_file_path, index=False)
```
