```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [0,1,2]
y = [0,3,3.5]

plt.errorbar(x, y, yerr=0.5, marker=".", markersize=20)
plt.xlabel("Level of Social Support")
plt.xticks(x)
plt.ylabel("Effect on Task Persistence")
```

```python
x = [0,1]
y = [1, 3]

plt.bar(x, y, color=["gray", "tab:green"])
plt.errorbar(x, y, yerr=0.5, fmt="o", color="black")
plt.ylabel("Task Persistence")
plt.xticks(x, labels=["Control", "Self-efficacy"])
```

```python
n1 = 4
n2 = 15
cluster_a = np.ones([n1,2]) + np.random.rand(n1,2)
cluster_b = np.ones([n1,2]) * 3 + np.random.rand(n1,2)
cluster_c = np.ones([n2,2]) @ np.array([[3, 0], [0, 1]]) + np.random.rand(n2,2)
cluster_d = np.ones([n2,2]) @ np.array([[1, 0], [0, 3]]) + np.random.rand(n2,2)

subjects = np.concatenate([cluster_a, cluster_b, cluster_c, cluster_d]).T

plt.scatter(*subjects, color="black")
plt.xlabel("Effect of Social Support")
plt.ylabel("Effect of Self-efficacy")
```
